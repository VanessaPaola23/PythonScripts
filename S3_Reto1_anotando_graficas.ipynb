```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

# Desactivar advertencias
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/content/drive/MyDrive/BEDU/Remoto Datasets/Remoto diabetes-clean.csv', index_col=0)

df.head()
```

### a)

```python
ax = sns.distplot(df['glucose'], kde=False, norm_hist=False)
ax.set(title= "Plasma glucose concentration a 2 hours in an oral glucose tolerance test", xlabel= 'Glucosa', ylabel= 'Conteo de personas');
```

### b)

```python
pregnancy_counts = df['pregnancies'].value_counts()
```

```python
print(pregnancy_counts)
```

```python
ax = sns.barplot(x=pregnancy_counts.index, y=pregnancy_counts)
ax.set(title= 'Cantidades de embarazos de las mujeres analizadas', xlabel= 'embarazos', ylabel= 'Conteo de Mujeres');
```

### c)

```python
# Filtro por saludables o enfermas
healthy = df[df['outcome'] == 0]
sick = df[df['outcome'] == 1]
```

```python
ax = sns.distplot(healthy['glucose'], hist=False, kde_kws={'shade': True}, label='Mujeres Sanas')
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True}, label='Mujeres Enfermas')
ax.set_title('Comparacion de glucosa de mujeres sanas o enfermas', fontsize= 20, pad= 20)
ax.set(xlabel= 'Glucosa', ylabel= 'Densidad relativa')
ax.legend(loc= 'upper right')
```

### d)

```python
ax = sns.distplot(sick['blood_pressure'], hist=False, kde_kws={'shade': True}, label= 'Presion')
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True}, label= 'Glucosa')
sns.distplot(sick['insulin'], hist=False, kde_kws={'shade': True}, label= 'Insulina')
ax.set_title('Representacion de mediciones en mujeres enfermas', fontsize= 20, pad= 20)
ax.set(xlabel= 'Valores', ylabel= 'Densidad relativa')
ax.legend(loc= 'upper right')
```

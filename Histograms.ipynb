```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, skew
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/beduExpert/B2-Analisis-de-Datos-con-Python-2020/master/Datasets/diabetes-clean.csv', index_col=0)
df.head()

df.dtypes #pregnancies, glucose, blood_pressure, skin_thickness, insulin,age
```

```python
# Graficar los resultados
sns.histplot(df['pregnancies'], bins=50, kde=False)

plt.title('Pregnancy Histogram')
plt.xlabel('Pregnancy')
plt.ylabel('Frecuencia')

# Mostrar la gráfica
plt.show()

# Calcular y mostrar curtosis y asimetría
skewness = skew(df['pregnancies'])
kurt = kurtosis(df['pregnancies'])

kurt, skewness
```

```python
# Graficar los resultados
sns.histplot(df['glucose'], bins=50, kde=False, color="purple")

plt.title('Glucose Histogram')
plt.xlabel('Glucose')
plt.ylabel('Frecuencia')

# Mostrar la gráfica
plt.show()

# Calcular y mostrar curtosis y asimetría
skewness = skew(df['glucose'])
kurt = kurtosis(df['glucose'])

kurt, skewness
```

```python
# Graficar los resultados
sns.histplot(df['blood_pressure'], bins=50, kde=False, color="magenta")

plt.title('blood_pressure Histogram')
plt.xlabel('blood_pressure')
plt.ylabel('Frecuencia')

# Mostrar la gráfica
plt.show()

# Calcular y mostrar curtosis y asimetría
skewness = skew(df['blood_pressure'])
kurt = kurtosis(df['blood_pressure'])

kurt, skewness
```

```python
# Graficar los resultados
sns.histplot(df['skin_thickness'], bins=50, kde=False, color="yellow")

plt.title('skin_thickness Histogram')
plt.xlabel('skin_thickness')
plt.ylabel('Frecuencia')

# Mostrar la gráfica
plt.show()

# Calcular y mostrar curtosis y asimetría
skewness = skew(df['skin_thickness'])
kurt = kurtosis(df['skin_thickness'])

kurt, skewness
```

```python
# Graficar los resultados
sns.histplot(df['insulin'], bins=50, kde=False, color="green")

plt.title('insulin Histogram')
plt.xlabel('insulin')
plt.ylabel('Frecuencia')

# Mostrar la gráfica
plt.show()

# Calcular y mostrar curtosis y asimetría
skewness = skew(df['insulin'])
kurt = kurtosis(df['insulin'])

kurt, skewness
```

```python
# Graficar los resultados
sns.histplot(df['age'], bins=50, kde=False,color="grey")

plt.title('age Histogram')
plt.xlabel('age')
plt.ylabel('Frecuencia')

# Mostrar la gráfica
plt.show()

# Calcular y mostrar curtosis y asimetría
skewness = skew(df['age'])
kurt = kurtosis(df['age'])

kurt, skewness
```
